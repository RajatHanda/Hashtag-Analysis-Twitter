```python
import pandas as pd
```

```python
df_desc=pd.read_csv("/Users/rajathanda/Downloads/ILAE_user_db.csv",sep=',', encoding=' ISO-8859-1')
```

```python
df_desc['description'].astype(str)
```

```python
def remove_non_ascii_1(df_desc):
    return ''.join([i if ord(i) < 128 else ' ' for i in df_desc])
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 1435):
    review = re.sub('[^a-zA-Z]', ' ', str(df_desc['description'][i]))
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
newlist2=pd.DataFrame(corpus)
```

```python
merged = pd.merge(left=df_desc, left_index=True,right=newlist2, right_index=True,how='inner')
```

```python
merged.to_csv("/Users/rajathanda/Downloads/ILAEText_Eng1.csv", sep=',', encoding='utf-8')
```
